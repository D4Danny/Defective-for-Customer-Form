```python
#Convert the discreet data into count and then apply CHI SQuARE formula

country1=caller['Phillippines'].value_counts()
country2=caller['Indonesia'].value_counts()
country3=caller['Malta'].value_counts()
country4=caller['India'].value_counts()
print(country1)
print(country2)
print(country3)
print(country4)
```

```python
#Chi-square test,INPUT(X) Discreet in TWO Categories and OUTPUT(Y) is also discreet 
chistats = sp.stats.chi2_contingency([[271,267,269,280],[29,33,31,20]])
(chistats[0], chistats[1])

if chistats[1] < 0.05:
  print('we reject null hypothesis')
else:
  print('we accept null hypothesis')
```
